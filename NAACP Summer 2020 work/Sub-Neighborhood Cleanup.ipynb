```python
Sub_neigh_list_clean = []
for names in Sub_neighborhoods_df['Primary Sub-Neighborhood']:
    if names in Neighborhoods_list:
        Sub_neigh_list_clean.append('')
    else:
        Sub_neigh_list_clean.append(names)

Sub_neighborhoods_df['Primary Sub-Neighborhood Cleaned'] = Sub_neigh_list_clean
Sub_neighborhoods_df
```

```python
Secondary_Sub_neigh_list_clean = []
for names in Sub_neighborhoods_df['Secondary Sub-Neighborhood']:
    if names in Neighborhoods_list:
        Secondary_Sub_neigh_list_clean.append('')
    else:
        Secondary_Sub_neigh_list_clean.append(names)

Sub_neighborhoods_df['Secondary Sub-Neighborhood Cleaned'] = Secondary_Sub_neigh_list_clean
Sub_neighborhoods_df
```

```python
Teritiary_Sub_neigh_list_clean = []
for names in Sub_neighborhoods_df['Teritriary Sub-Neighborhood(s)']:
    if names in Neighborhoods_list:
        Teritiary_Sub_neigh_list_clean.append('')
    else:
        Teritiary_Sub_neigh_list_clean.append(names)

Sub_neighborhoods_df['Teritiary Sub-Neighborhood(s) Cleaned'] = Teritiary_Sub_neigh_list_clean
Sub_neighborhoods_df
```
